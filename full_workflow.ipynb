```python
# Extract all available classes of service from search response
try:
    search_response
except NameError:
    with open("search_response.json") as f:
        search_response = json.load(f)

# Get all products and extract unique class of service + cabin combinations
products = search_response["CatalogProductOfferingsResponse"]["ReferenceList"][1]["Product"]

class_of_service_options = {}

for product in products:
    if "PassengerFlight" in product and len(product["PassengerFlight"]) > 0:
        for passenger_flight in product["PassengerFlight"]:
            if "FlightProduct" in passenger_flight and len(passenger_flight["FlightProduct"]) > 0:
                for flight_product in passenger_flight["FlightProduct"]:
                    cos = flight_product.get("classOfService", "N/A")
                    cabin = flight_product.get("cabin", "N/A")
                    fare_basis = flight_product.get("fareBasisCode", "N/A")
                    
                    key = f"{cos} - {cabin}"
                    if key not in class_of_service_options:
                        class_of_service_options[key] = {
                            "classOfService": cos,
                            "cabin": cabin,
                            "fareBasisCodes": [],
                            "products": []
                        }
                    if fare_basis not in class_of_service_options[key]["fareBasisCodes"]:
                        class_of_service_options[key]["fareBasisCodes"].append(fare_basis)
                    if product.get("id") not in class_of_service_options[key]["products"]:
                        class_of_service_options[key]["products"].append(product.get("id"))

print("Available Classes of Service and Cabins:\n")
print("=" * 60)
for key, info in sorted(class_of_service_options.items()):
    print(f"Class of Service: {info['classOfService']}")
    print(f"Cabin: {info['cabin']}")
    print(f"Available in {len(info['products'])} product(s): {', '.join(info['products'][:5])}")
    if len(info['products']) > 5:
        print(f"  ... and {len(info['products']) - 5} more")
    print(f"Fare Basis Codes: {', '.join(info['fareBasisCodes'][:3])}")
    if len(info['fareBasisCodes']) > 3:
        print(f"  ... and {len(info['fareBasisCodes']) - 3} more")
    print("-" * 60)
```
